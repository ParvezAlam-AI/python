## Creating Pandas DataFrame

```python
population_dict = {
    'California':38956785,
    'Texas':26441568,
    'New York':19555647,
    'Florida':12364569,
    'Illinois':12882135
}

population = pd.Series(population_dict)
df = pd.DataFrame(population)
df
```

```python
#creating the pandas Datafrome with list of the dictionaries
data = [{'a':i,'b':2*i} for i in range(3)]
data
```

```python
df = pd.DataFrame(data)
df
```

```python
# a customized index can be setup for a smaller DataFrame
df = pd.DataFrame(data,index = ['x','y','z'])
df
```

```python
df.columns
```

```python
# change the name of the column
# to change the name of the columns use 'df.column = list of the new column names'
df.columns=['A','B']
df
```

```python
#if a DataFrame is created using the dictionary and the key are not matched in two dictionaries then it will be filled with NaN values

check_nan = pd.DataFrame([{'a':1,'b':2},{'b':3,'c':4}])
check_nan
```

```python
check_nan = pd.DataFrame([{'a':1,'b':2},{'b':3,'c':4}])
adddata = [{'a':3,'e':6}]
adddata = pd.DataFrame(adddata)
new_dict = pd.concat([check_nan,adddata],ignore_index = True)
new_dict
```
